## **Experiment**

This pipeline processes the file output from the DSSP Server (https://pdb-redo.eu/dssp), which has annotated an input .pdb file containing the first 193 amino acids of E. coli DcuS. 

The following Python packages must be installed prior to loading into the Jupyter Notebook session. See the Reproducibility section below for a complete list of packages and their versions used in this workflow. 

```python
from mkdssp import dssp
import numpy as np
import pandas as pd
```

Import the annotated DSSP file from the DSSP Server (detailed methods are provided in the Methods section "Secondary structure annotation")

```python
# Read in the file
file_path = "nohd.cif"

with open(file_path, "rt") as f:
    file_content = f.read()

dcus = dssp(file_content)

# Print out the data inside
print("residues: ", dcus.statistics.residues)

# Re-write the data as a Pandas Dataframe
seq_id, compound_id, compound_letter, structure = [], [], [], []

for res in dcus:
    print(res.asym_id, res.seq_id, res.compound_id, res.compound_letter, res.type)
    seq_id.append(res.seq_id)
    compound_id.append(res.compound_id)
    compound_letter.append(res.compound_letter)
    structure.append(str(res.type))
    
dcus_df = pd.DataFrame()
dcus_df["seq_id"] = seq_id
dcus_df["compound_id"] = compound_id
dcus_df["compound_letter"] = compound_letter
dcus_df["structure"] = structure
```

Export the data as a .CSV file to load into an R session. 

```python
dcus_df.to_csv("DcuS_secondary_structure_DSSP.csv",index=False)
```
